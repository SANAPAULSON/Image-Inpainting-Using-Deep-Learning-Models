```python
import os
import numpy as np
import tensorflow as tf
import pickle
import matplotlib.pyplot as plt
from tensorflow.keras.layers import (Input, Conv2D, MaxPooling2D, UpSampling2D,
                                     BatchNormalization, Activation, Add, Conv2DTranspose, Concatenate)
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing.image import img_to_array, load_img
import os
import tensorflow as tf


def load_images_from_directory(directory, target_size=(128, 128)):
    images = []
    for filename in os.listdir(directory):
        filepath = os.path.join(directory, filename)
        img = load_img(filepath, target_size=target_size)
        img_array = img_to_array(img) / 255.0
        images.append(img_array)
    return np.array(images)

# Load dataset
train_dir = "prepared_data1/train"
val_dir = "prepared_data1/val"
test_dir = "prepared_data1/test"

# x_train = load_images_from_directory(train_dir)
# x_val = load_images_from_directory(val_dir)
x_test = load_images_from_directory(test_dir)

# # Save and reload dataset as pickle
# with open('x_train.pkl', 'wb') as f:
#     pickle.dump(x_train, f)
# with open('x_val.pkl', 'wb') as f:
#     pickle.dump(x_val, f)
with open('x_test.pkl', 'wb') as f:
    pickle.dump(x_test, f)

with open('x_train.pkl', 'rb') as f:
    x_train = pickle.load(f)
with open('x_val.pkl', 'rb') as f:
    x_val = pickle.load(f)
with open('x_test.pkl', 'rb') as f:
    x_test = pickle.load(f)


noise_factor = 0.2
x_train_noisy = np.clip(x_train + noise_factor * np.random.normal(size=x_train.shape), 0.0, 1.0)
x_val_noisy = np.clip(x_val + noise_factor * np.random.normal(size=x_val.shape), 0.0, 1.0)
x_test_noisy = np.clip(x_test + noise_factor * np.random.normal(size=x_test.shape), 0.0, 1.0)

def residual_block(x, filters):
    res = Conv2D(filters, (3, 3), padding="same", activation="relu")(x)
    res = Conv2D(filters, (3, 3), padding="same", activation="relu")(res)

    if x.shape[-1] != filters:
        x = Conv2D(filters, (1, 1), padding="same")(x)

    return Add()([x, res])

def build_ddpm(input_shape=(128, 128, 3)):
    input_img = Input(shape=input_shape)

    e1 = Conv2D(64, (3, 3), padding='same', activation='relu')(input_img)
    e1 = residual_block(e1, 64)
    p1 = MaxPooling2D((2, 2), padding='same')(e1)

    e2 = Conv2D(128, (3, 3), padding='same', activation='relu')(p1)
    e2 = residual_block(e2, 128)
    p2 = MaxPooling2D((2, 2), padding='same')(e2)

    e3 = Conv2D(256, (3, 3), padding='same', activation='relu')(p2)
    e3 = residual_block(e3, 256)
    p3 = MaxPooling2D((2, 2), padding='same')(e3)

    e4 = Conv2D(512, (3, 3), padding='same', activation='relu')(p3)
    e4 = residual_block(e4, 512)
    p4 = MaxPooling2D((2, 2), padding='same')(e4)

    b = Conv2D(1024, (3, 3), padding='same', activation='relu')(p4)
    b = residual_block(b, 1024)

    d4 = Conv2DTranspose(512, (3, 3), strides=(2, 2), padding='same', activation='relu')(b)
    d4 = Concatenate()([d4, e4])
    d4 = residual_block(d4, 512)

    d3 = Conv2DTranspose(256, (3, 3), strides=(2, 2), padding='same', activation='relu')(d4)
    d3 = Concatenate()([d3, e3])
    d3 = residual_block(d3, 256)

    d2 = Conv2DTranspose(128, (3, 3), strides=(2, 2), padding='same', activation='relu')(d3)
    d2 = Concatenate()([d2, e2])
    d2 = residual_block(d2, 128)

    d1 = Conv2DTranspose(64, (3, 3), strides=(2, 2), padding='same', activation='relu')(d2)
    d1 = Concatenate()([d1, e1])
    d1 = residual_block(d1, 64)

    decoded = Conv2D(3, (3, 3), activation='sigmoid', padding='same')(d1)

    return Model(input_img, decoded, name="DDPM_Like_Autoencoder")

ddpm_like_autoencoder = build_ddpm_like_autoencoder()
ddpm_like_autoencoder.compile(optimizer='adam', loss='mse')

ddpm_like_autoencoder.summary()

history = ddpm_like_autoencoder.fit(
    x_train_noisy, x_train,
    epochs=200,
    batch_size=32,
    shuffle=True,
    validation_data=(x_val_noisy, x_val)
)

ddpm_like_autoencoder.save("ddpm_denoising.h5")

plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.legend()
plt.title("Training History")
plt.show()

decoded_imgs = ddpm_like_autoencoder.predict(x_test_noisy)

n = 5
plt.figure(figsize=(20, 6))
for i in range(n):
    ax = plt.subplot(3, n, i + 1)
    plt.imshow(x_test_noisy[i])
    plt.title("Noisy Image")
    plt.axis("off")

    ax = plt.subplot(3, n, i + n + 1)
    plt.imshow(decoded_imgs[i])
    plt.title("Denoised Image")
    plt.axis("off")

    ax = plt.subplot(3, n, i + 2 * n + 1)
    plt.imshow(x_test[i])
    plt.title("Original Image")
    plt.axis("off")
plt.show()
```
