```python
import rarfile

# Path to the .rar file
rar_file = 'celeba_hq_256.rar'

# Path where you want to extract the contents
extract_to = 'Dataset'

# Open the .rar file
with rarfile.RarFile(rar_file) as rf:
    rf.extractall(path=extract_to)

print(f"Extraction complete! Files extracted to {extract_to}")
```
