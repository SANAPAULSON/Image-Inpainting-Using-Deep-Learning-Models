```python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import (Input, Conv2D, MaxPooling2D, UpSampling2D, BatchNormalization,
                                     LeakyReLU, Flatten, Dense, Dropout, Activation, Conv2DTranspose, Concatenate)
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt

import pickle
with open('x_train.pkl', 'rb') as f:
    x_train = pickle.load(f)
with open('x_val.pkl', 'rb') as f:
    x_val = pickle.load(f)

noise_factor = 0.2
x_train_noisy = np.clip(x_train + noise_factor * np.random.normal(size=x_train.shape), 0.0, 1.0)
x_val_noisy = np.clip(x_val + noise_factor * np.random.normal(size=x_val.shape), 0.0, 1.0)


```

```python
from tensorflow.keras.layers import (Input, Conv2D, MaxPooling2D, UpSampling2D, BatchNormalization,
                                     LeakyReLU, Flatten, Dense, Dropout, Activation, Conv2DTranspose,
                                     Concatenate, Add, GlobalAveragePooling2D)
def residual_block(x, filters):
    res = Conv2D(filters, (3, 3), padding='same')(x)
    res = BatchNormalization()(res)
    res = Activation('relu')(res)
    res = Conv2D(filters, (3, 3), padding='same')(res)
    res = BatchNormalization()(res)
    return Add()([x, res])

def build_generator():
    inputs = Input(shape=(128, 128, 3))

    e1 = Conv2D(64, (3, 3), padding='same', activation='relu')(inputs)
    e1 = BatchNormalization()(e1)
    e1_pool = MaxPooling2D((2, 2))(e1)

    e2 = Conv2D(128, (3, 3), padding='same', activation='relu')(e1_pool)
    e2 = BatchNormalization()(e2)
    e2_pool = MaxPooling2D((2, 2))(e2)

    b = Conv2D(256, (3, 3), padding='same', activation='relu')(e2_pool)
    b = BatchNormalization()(b)
    b = residual_block(b, 256)

    d1 = UpSampling2D((2, 2))(b)
    d1 = Conv2DTranspose(128, (3, 3), padding='same', activation='relu')(d1)
    d1 = BatchNormalization()(d1)
    d1 = Concatenate()([d1, e2])

    d2 = UpSampling2D((2, 2))(d1)
    d2 = Conv2DTranspose(64, (3, 3), padding='same', activation='relu')(d2)
    d2 = BatchNormalization()(d2)
    d2 = Concatenate()([d2, e1])

    outputs = Conv2D(3, (3, 3), activation='sigmoid', padding='same')(d2)

    return Model(inputs, outputs, name="Generator")

generator = build_generator()

def build_discriminator():
    inputs = Input(shape=(128, 128, 3))
    x = Conv2D(64, (4, 4), strides=2, padding='same')(inputs)
    x = LeakyReLU(alpha=0.2)(x)

    x = Conv2D(128, (4, 4), strides=2, padding='same')(x)
    x = BatchNormalization()(x)
    x = LeakyReLU(alpha=0.2)(x)

    x = Conv2D(256, (4, 4), strides=2, padding='same')(x)
    x = BatchNormalization()(x)
    x = LeakyReLU(alpha=0.2)(x)

    x = GlobalAveragePooling2D()(x)
    x = Dense(1, activation='sigmoid')(x)

    model = Model(inputs, x)
    model.compile(optimizer=Adam(learning_rate=0.0002, beta_1=0.5), loss='binary_crossentropy', metrics=['accuracy'])
    return model

discriminator = build_discriminator()

# ------------------------------
# GAN Model (Adversarial + L1 Loss)
# ------------------------------
def build_gan(generator, discriminator):
    discriminator.trainable = False
    inputs = Input(shape=(128, 128, 3))
    generated_images = generator(inputs)
    validity = discriminator(generated_images)

    gan = Model(inputs, [generated_images, validity])

    gan.compile(
        optimizer=Adam(learning_rate=0.0002, beta_1=0.5),
        loss=['mse', 'binary_crossentropy'],
        loss_weights=[100, 1]
    )

    return gan

gan = build_gan(generator, discriminator)

# ------------------------------
# Training Loop
# ------------------------------
epochs = 5000
batch_size = 32

for epoch in range(epochs):
    idx = np.random.randint(0, x_train.shape[0], batch_size // 2)
    real_images = x_train[idx]
    noisy_images = x_train_noisy[idx]
    generated_images = generator.predict(noisy_images)

    real_labels = np.ones((batch_size // 2, 1))
    fake_labels = np.zeros((batch_size // 2, 1))

    d_loss_real = discriminator.train_on_batch(real_images, real_labels)
    d_loss_fake = discriminator.train_on_batch(generated_images, fake_labels)
    d_loss = 0.5 * np.add(d_loss_real, d_loss_fake)

    idx = np.random.randint(0, x_train_noisy.shape[0], batch_size)
    noise_batch = x_train_noisy[idx]
    valid_labels = np.ones((batch_size, 1))

    g_loss = gan.train_on_batch(noise_batch, [x_train[idx], valid_labels])

    print(f"Epoch {epoch+1}/{epochs} | D Loss: {d_loss[0]:.4f}, D Acc: {d_loss[1]:.4f} | G Loss: {g_loss[0]:.4f}")

    if epoch % 10 == 0:
        generator.save("denoising_gan_generator.h5")

# ------------------------------
# Testing
# ------------------------------
def load_images(directory, target_size=(128, 128)):
    images = []
    for filename in os.listdir(directory):
        img = tf.keras.preprocessing.image.load_img(os.path.join(directory, filename), target_size=target_size)
        images.append(tf.keras.preprocessing.image.img_to_array(img) / 255.0)
    return np.array(images)

x_test = load_images("prepared_data/test")
x_test_noisy = np.clip(x_test + noise_factor * np.random.normal(size=x_test.shape), 0.0, 1.0)

denoised_images = generator.predict(x_test_noisy)

plt.imshow(denoised_images[0])
plt.show()
```
