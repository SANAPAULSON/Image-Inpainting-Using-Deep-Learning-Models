```python
import os
import numpy as np
import tensorflow as tf
import pickle
import matplotlib.pyplot as plt
from tensorflow.keras.layers import (Input, Conv2D, MaxPooling2D, UpSampling2D,
                                     BatchNormalization, Activation, Add, Conv2DTranspose, Concatenate)
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing.image import img_to_array, load_img

def load_images_from_directory(directory, max_images, target_size=(128, 128)):
    images = []
    filenames = os.listdir(directory)[:max_images]  # Limit image count
    for filename in filenames:
        filepath = os.path.join(directory, filename)
        img = load_img(filepath, target_size=target_size)
        img_array = img_to_array(img) / 255.0
        images.append(img_array)
    return np.array(images)

# Dataset Paths
train_dir = "prepared_data1/train"
val_dir = "prepared_data1/val"
test_dir = "prepared_data1/test"

# Load limited dataset
x_train = load_images_from_directory(train_dir, max_images=1000)
x_val = load_images_from_directory(val_dir, max_images=100)
x_test = load_images_from_directory(test_dir, max_images=100)

# Save dataset for reuse
with open('x_train1.pkl', 'wb') as f:
    pickle.dump(x_train, f)
with open('x_val1.pkl', 'wb') as f:
    pickle.dump(x_val, f)
with open('x_test1.pkl', 'wb') as f:
    pickle.dump(x_test, f)

# Load dataset from pickle
with open('x_train1.pkl', 'rb') as f:
    x_train = pickle.load(f)
with open('x_val1.pkl', 'rb') as f:
    x_val = pickle.load(f)
with open('x_test1.pkl', 'rb') as f:
    x_test = pickle.load(f)

# Function to Create Masked Images
def mask_images(images, mask_size=(32, 32)):
    masked_images = images.copy()
    masks = np.ones_like(images)

    for img, mask in zip(masked_images, masks):
        h, w, _ = img.shape
        x = np.random.randint(0, h - mask_size[0])
        y = np.random.randint(0, w - mask_size[1])
        img[x:x + mask_size[0], y:y + mask_size[1], :] = 0  # Mask with black patch
        mask[x:x + mask_size[0], y:y + mask_size[1], :] = 0  # Store mask position

    return np.array(masked_images), np.array(masks)

# Create Masked Datasets
x_train_masked, x_train_masks = mask_images(x_train)
x_val_masked, x_val_masks = mask_images(x_val)
x_test_masked, x_test_masks = mask_images(x_test)

# Residual Block
def residual_block(x, filters):
    res = Conv2D(filters, (3, 3), padding="same", activation="relu")(x)
    res = Conv2D(filters, (3, 3), padding="same", activation="relu")(res)
    if x.shape[-1] != filters:
        x = Conv2D(filters, (1, 1), padding="same")(x)
    return Add()([x, res])

# Inpainting Model
def build_inpainting_model(input_shape=(128, 128, 3)):
    input_img = Input(shape=input_shape)
    input_mask = Input(shape=input_shape)
    merged_input = Concatenate()([input_img, input_mask])

    e1 = Conv2D(64, (3, 3), padding='same', activation='relu')(merged_input)
    e1 = residual_block(e1, 64)
    p1 = MaxPooling2D((2, 2), padding='same')(e1)

    e2 = Conv2D(128, (3, 3), padding='same', activation='relu')(p1)
    e2 = residual_block(e2, 128)
    p2 = MaxPooling2D((2, 2), padding='same')(e2)

    e3 = Conv2D(256, (3, 3), padding='same', activation='relu')(p2)
    e3 = residual_block(e3, 256)
    p3 = MaxPooling2D((2, 2), padding='same')(e3)

    e4 = Conv2D(512, (3, 3), padding='same', activation='relu')(p3)
    e4 = residual_block(e4, 512)
    p4 = MaxPooling2D((2, 2), padding='same')(e4)

    b = Conv2D(1024, (3, 3), padding='same', activation='relu')(p4)
    b = residual_block(b, 1024)

    d4 = Conv2DTranspose(512, (3, 3), strides=(2, 2), padding='same', activation='relu')(b)
    d4 = Concatenate()([d4, e4])
    d4 = residual_block(d4, 512)

    d3 = Conv2DTranspose(256, (3, 3), strides=(2, 2), padding='same', activation='relu')(d4)
    d3 = Concatenate()([d3, e3])
    d3 = residual_block(d3, 256)

    d2 = Conv2DTranspose(128, (3, 3), strides=(2, 2), padding='same', activation='relu')(d3)
    d2 = Concatenate()([d2, e2])
    d2 = residual_block(d2, 128)

    d1 = Conv2DTranspose(64, (3, 3), strides=(2, 2), padding='same', activation='relu')(d2)
    d1 = Concatenate()([d1, e1])
    d1 = residual_block(d1, 64)

    decoded = Conv2D(3, (3, 3), activation='sigmoid', padding='same')(d1)

    return Model([input_img, input_mask], decoded, name="Image_Inpainting_Model")

# Compile Model
inpainting_model = build_inpainting_model()
inpainting_model.compile(optimizer='adam', loss='mse')

# Train the Model
history = inpainting_model.fit(
    [x_train_masked, x_train_masks], x_train,
    epochs=100,
    batch_size=32,
    shuffle=True,
    validation_data=([x_val_masked, x_val_masks], x_val)
)

# Save Model
inpainting_model.save("image_inpainting.h5")

# Plot Training History
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.legend()
plt.title("Training History")
plt.show()
```

```python
import numpy as np
import tensorflow as tf
import pickle
import matplotlib.pyplot as plt

# Load the trained model
model = tf.keras.models.load_model("image_inpainting.h5", custom_objects={'mse': tf.keras.losses.MeanSquaredError()})

# Load the test dataset
with open('x_test1.pkl', 'rb') as f:
    x_test = pickle.load(f)

# Function to Create Masked Images (Same as used during training)
def mask_images(images, mask_size=(32, 32)):
    masked_images = images.copy()
    masks = np.ones_like(images)

    for img, mask in zip(masked_images, masks):
        h, w, _ = img.shape
        x = np.random.randint(0, h - mask_size[0])
        y = np.random.randint(0, w - mask_size[1])
        img[x:x + mask_size[0], y:y + mask_size[1], :] = 0  # Apply black mask
        mask[x:x + mask_size[0], y:y + mask_size[1], :] = 0  # Store mask location

    return np.array(masked_images), np.array(masks)

# Create Masked Test Dataset
x_test_masked, x_test_masks = mask_images(x_test)

# Predict on the Test Data
predicted_images = model.predict([x_test_masked, x_test_masks])

# Select 5 Random Indices to Display
random_indices = np.random.choice(len(x_test), 5, replace=False)

# Plot Results
fig, axes = plt.subplots(5, 3, figsize=(10, 15))

for i, idx in enumerate(random_indices):
    # Original, Masked, and Inpainted Images
    original = x_test[idx]
    masked = x_test_masked[idx]
    inpainted = predicted_images[idx]

    # Display Original Image
    axes[i, 0].imshow(original)
    axes[i, 0].set_title("Original")
    axes[i, 0].axis("off")

    # Display Masked Image
    axes[i, 1].imshow(masked)
    axes[i, 1].set_title("Masked")
    axes[i, 1].axis("off")

    # Display Reconstructed Image
    axes[i, 2].imshow(inpainted)
    axes[i, 2].set_title("Inpainted")
    axes[i, 2].axis("off")

plt.tight_layout()
plt.show()
```
